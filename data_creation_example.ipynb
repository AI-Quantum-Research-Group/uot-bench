### Using datasets of images

```python
from uot.experiment import generate_data_problems
from uot.suites import time_precision_suite
from algorithms.sinkhorn import sinkhorn
from algorithms.lp import pot_lp
from algorithms.pdgd import pdhg_algorithm as pdhg
```

```python
cauchy_problems = generate_data_problems(data_type="CauchyDensity", num_points=32, num_samples=2)
classic_images_problems = generate_data_problems(data_type="ClassicImages", num_points=32, num_samples=2)

all_problems = cauchy_problems + classic_images_problems
```

```python
solvers = {
    "Sinkhorn": sinkhorn,
    "Linear Program": pot_lp,
    "PDHG": pdhg
}

results = {}
for solver_name, solver in solvers.items():
    print(f"Running {solver_name}...")
    result = time_precision_suite.run_suite(
        name=solver_name,
        ot_problems=all_problems,
        solver=solver
    )
    results[solver_name] = result
    
    print(f"\nFinished {solver_name}.\n")
```

### Using 3D Colored Meshes

```python
from uot.experiment import generate_3d_mesh_problems
import matplotlib.pyplot as plt
```

```python
red_channel_problems = generate_3d_mesh_problems(num_points=1000, num_samples=3, color_channel=0)
print(f"Generated {len(red_channel_problems)} problems using the red channel")

green_channel_problems = generate_3d_mesh_problems(num_points=1000, num_samples=3, color_channel=1)
print(f"Generated {len(green_channel_problems)} problems using the green channel")

blue_channel_problems = generate_3d_mesh_problems(num_points=1000, num_samples=3, color_channel=2)
print(f"Generated {len(blue_channel_problems)} problems using the blue channel")
```

```python
if len(red_channel_problems) > 0:
    problem = red_channel_problems[0]
    
    print(f"Problem: {problem.name}")
    print(f"Source measure: {problem.source_measure.name}")
    print(f"Target measure: {problem.target_measure.name}")
    print(f"\nSource mesh: {problem.source_measure.kwargs['mesh_name']}")
    print(f"Target mesh: {problem.target_measure.kwargs['mesh_name']}")
    
    print(f"\nSource distribution size: {problem.source_measure.distribution.shape}")
    print(f"Target distribution size: {problem.target_measure.distribution.shape}")
    
    plt.figure(figsize=(12, 5))
    
    plt.subplot(1, 2, 1)
    plt.hist(problem.source_measure.distribution, bins=30, alpha=0.7)
    plt.title(f"Source: {problem.source_measure.name}")
    plt.xlabel("Color value (red channel)")
    plt.ylabel("Frequency")
    
    plt.subplot(1, 2, 2)
    plt.hist(problem.target_measure.distribution, bins=30, alpha=0.7)
    plt.title(f"Target: {problem.target_measure.name}")
    plt.xlabel("Color value (red channel)")
    plt.ylabel("Frequency")
    
    plt.tight_layout()
    plt.show()
```

```python
if len(red_channel_problems) > 0:
    from uot.suites import time_precision_suite
    from algorithms.sinkhorn import sinkhorn
    from algorithms.lp import pot_lp
    
    mesh_problems = red_channel_problems[:3]
    
    solvers = {
        "Sinkhorn": sinkhorn,
        "Linear Program": pot_lp
    }
    
    mesh_results = {}
    for solver_name, solver in solvers.items():
        print(f"Running {solver_name} on 3D mesh problems...")
        result = time_precision_suite.run_suite(
            name=solver_name,
            ot_problems=mesh_problems,
            solver=solver
        )
        mesh_results[solver_name] = result
        
        print(f"\nFinished {solver_name}.\n")
```
