### Using datasets of images

```python
from uot.experiment import generate_data_problems
from uot.suites import time_precision_suite
from algorithms.sinkhorn import sinkhorn
from algorithms.lp import pot_lp
from algorithms.pdgd import pdhg_algorithm as pdhg
```

```python
cauchy_problems = generate_data_problems(data_type="CauchyDensity", num_points=32, num_samples=2)
classic_images_problems = generate_data_problems(data_type="ClassicImages", num_points=32, num_samples=2)

all_problems = cauchy_problems + classic_images_problems
```

```python
solvers = {
    "Sinkhorn": sinkhorn,
    "Linear Program": pot_lp,
    "PDHG": pdhg
}

results = {}
for solver_name, solver in solvers.items():
    print(f"Running {solver_name}...")
    result = time_precision_suite.run_suite(
        name=solver_name,
        ot_problems=all_problems,
        solver=solver
    )
    results[solver_name] = result
    
    print(f"\nFinished {solver_name}.\n")
```

### Using 3D Colored Meshes

```python
from uot.experiment import generate_3d_mesh_problems
import matplotlib.pyplot as plt
import numpy as np
import os
from uot.suites import time_precision_suite
from algorithms.sinkhorn import sinkhorn
from algorithms.lp import pot_lp
    
```

```python
rgb_problems = generate_3d_mesh_problems(num_points=1000, num_samples=3, color_mode="rgb_avg")
print(f"Generated {len(rgb_problems)} RGB combined problems")

red_problems = generate_3d_mesh_problems(num_points=1000, num_samples=3, color_mode="r")
print(f"Generated {len(red_problems)} red channel problems")

green_problems = generate_3d_mesh_problems(num_points=1000, num_samples=3, color_mode="g")
print(f"Generated {len(green_problems)} green channel problems")

blue_problems = generate_3d_mesh_problems(num_points=1000, num_samples=3, color_mode="b")
print(f"Generated {len(blue_problems)} blue channel problems")

all_channel_problems = generate_3d_mesh_problems(num_points=1000, num_samples=3, color_mode="separate")
print(f"Generated {len(all_channel_problems)} problems across all color channels")
```

```python
if len(rgb_problems) > 0:
    problem = rgb_problems[0]
    
    print(f"Problem: {problem.name}")
    print(f"Source measure: {problem.source_measure.name}")
    print(f"Target measure: {problem.target_measure.name}")
    print(f"\nSource mesh: {problem.source_measure.kwargs['mesh_name']}")
    print(f"Target mesh: {problem.target_measure.kwargs['mesh_name']}")
    print(f"Color mode: {problem.source_measure.kwargs['color_mode']}")
    
    print(f"\nSource points: {problem.source_measure.kwargs['num_points']}")
    print(f"Target points: {problem.target_measure.kwargs['num_points']}")
    
    plt.figure(figsize=(12, 5))
    
    plt.subplot(1, 2, 1)
    plt.hist(problem.source_measure.distribution, bins=30, alpha=0.7)
    plt.title(f"Source: {problem.source_measure.name}")
    plt.xlabel("Color value (RGB average)")
    plt.ylabel("Frequency")
    
    plt.subplot(1, 2, 2)
    plt.hist(problem.target_measure.distribution, bins=30, alpha=0.7)
    plt.title(f"Target: {problem.target_measure.name}")
    plt.xlabel("Color value (RGB average)")
    plt.ylabel("Frequency")
    
    plt.tight_layout()
    plt.show()
```

```python
if len(red_problems) > 0 and len(green_problems) > 0 and len(blue_problems) > 0 and len(rgb_problems) > 0:

    red_source_mesh = red_problems[0].source_measure.kwargs['mesh_name']
    
    red_dist = red_problems[0].source_measure.distribution
    green_dist = green_problems[0].source_measure.distribution
    blue_dist = blue_problems[0].source_measure.distribution
    rgb_dist = rgb_problems[0].source_measure.distribution
    
    plt.figure(figsize=(15, 8))
    
    plt.subplot(2, 2, 1)
    plt.hist(red_dist, bins=30, color='red', alpha=0.7)
    plt.title(f"Red Channel - {red_source_mesh}")
    plt.xlabel("Red value")
    plt.ylabel("Frequency")
    
    plt.subplot(2, 2, 2)
    plt.hist(green_dist, bins=30, color='green', alpha=0.7)
    plt.title(f"Green Channel - {red_source_mesh}")
    plt.xlabel("Green value")
    plt.ylabel("Frequency")
    
    plt.subplot(2, 2, 3)
    plt.hist(blue_dist, bins=30, color='blue', alpha=0.7)
    plt.title(f"Blue Channel - {red_source_mesh}")
    plt.xlabel("Blue value")
    plt.ylabel("Frequency")
    
    plt.subplot(2, 2, 4)
    plt.hist(rgb_dist, bins=30, color='purple', alpha=0.7)
    plt.title(f"RGB Combined - {red_source_mesh}")
    plt.xlabel("RGB average value")
    plt.ylabel("Frequency")
    
    plt.tight_layout()
    plt.show()
```

```python
if len(rgb_problems) > 0:
    from uot.suites import time_precision_suite
    from algorithms.sinkhorn import sinkhorn
    from algorithms.lp import pot_lp
    
    test_problem = rgb_problems[0:1]
    
    print(f"Running solvers on problem: {test_problem[0].name}")
    print(f"Source points: {test_problem[0].source_measure.kwargs['num_points']}")
    print(f"Target points: {test_problem[0].target_measure.kwargs['num_points']}")
    
    solvers = {
        "Sinkhorn": sinkhorn
    }
    
    mesh_results = {}
    for solver_name, solver in solvers.items():
        print(f"Running {solver_name} on 3D mesh problem...")
        result = time_precision_suite.run_suite(
            name=solver_name,
            ot_problems=test_problem,
            solver=solver
        )
        mesh_results[solver_name] = result
        
        print(f"\nFinished {solver_name}.\n")
        
        result.display_result()
```
